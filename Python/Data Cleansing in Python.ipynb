# Outlier Treatment

```python
import seaborn as sns
sns.get_dataset_names()
```

```python
df1 = sns.load_dataset('car_crashes')
df1.head()
```

```python
sns.boxplot(df1['alcohol'])
```

```python
q1 = df1['alcohol'].quantile(0.25)
q3 = df1['alcohol'].quantile(0.75)
iqr = q3 - q1 
iqr
```

```python
df1['alcohol'] = df1['alcohol'][~((df1['alcohol'] < (q1 - 1.5*iqr))
                                  |(df1['alcohol'] > (q3 + 1.5*iqr)))]
```

```python
sns.boxplot(df1['alcohol'])
```

# Simple Imputer in Python

```python
import pandas as pd
from sklearn.impute import SimpleImputer
```

```python
df = pd.read_csv("unclean_data.csv",encoding='latin1')
df.head()
```

### strategy parameter in simpleimputer

If “mean”, then replace missing values using the mean along each column. Can only be used with numeric data.

If “median”, then replace missing values using the median along each column. Can only be used with numeric data.

If “most_frequent”, then replace missing using the most frequent value along each column. Can be used with strings or numeric data. If there is more than one such value, only the smallest is returned.

If “constant”, then replace missing values with fill_value. Can be used with strings or numeric data.

```python
import numpy as np
imputing = SimpleImputer(missing_values = np.nan, strategy = 'most_frequent')
rep = imputing.fit_transform(df)
rep1 = pd.DataFrame(rep)
rep1
```

## Bad Character Removal using Replace/Slice

```python
import pandas as pd
```

```python
data = ['ϕAashirϕ' , 'ϕEshanϕ', 'ϕKishoreϕ', 'ϕPriyanshiϕ', 'ϕShreyaϕ', 'ϕYashϕ']
data
```

```python
df = pd.DataFrame(data, columns = ['Names'])
df
```

```python
df['Names'] = df.Names.str.replace('ϕ', '')
df
```

```python
data = ['ϕAashirϕ' , 'ϕEshanϕ', 'ϕKishoreϕ', 'ϕPriyanshiϕ', 'ϕShreyaϕ', 'ϕYashϕ']
df = pd.DataFrame(data, columns = ['Names'])
print(df)
print('=============================================================================')
df['Names'] = df.Names.str.slice(1, -1)
print(df)
```

### Removing multiple bad characters using replace

```python
data = ['12Aashirϕß' , 'ϕ45Eshßanϕ', 'ϕ56Kißshoreϕ', 'ϕ102Priyßanshiϕß', 'ϕ398Sßhreyaßϕ', 'ϕß99999Yashϕ']
df = pd.DataFrame(data, columns = ['Names'])
print(df)
print('=============================================================================')
df['Names'] = df.Names.str.replace('ϕ', '').str.replace('ß', '').str.replace('1', '').str.replace('2',''). \
                str.replace('3', '').str.replace('4', '').str.replace('5', '').str.replace('6',''). \
                str.replace('0', '').str.replace('9', '').str.replace('8', '')
print(df)
```
