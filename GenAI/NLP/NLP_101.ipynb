### Tokenization

Tokenization is the process of splitting text into smaller units, called tokens. These tokens can be words, phrases, or even sentences, and are essential for many natural language processing (NLP) tasks.

#### Types of Tokenization:

1. **Word Tokenization**:
   - Word tokenization divides a text into individual words (tokens).
   - For example, the sentence `"The quick brown fox"` becomes the tokens `["The", "quick", "brown", "fox"]`.
   - It's useful for tasks like text classification, information retrieval, and machine learning, where analyzing individual words is important.

2. **Sentence Tokenization**:
   - Sentence tokenization breaks down a text into individual sentences.
   - For example, the text `"Hello world! How are you?"` becomes `["Hello world!", "How are you?"]`.
   - It helps in summarization, text analysis, and understanding the structure of the content.

#### Importance of Tokenization:
- Tokenization is a fundamental step in natural language processing because it transforms raw text into meaningful chunks that can be processed by machines.
- It plays a crucial role in preparing text for further analysis, like stemming, lemmatization, or part-of-speech tagging.

### Porter Stemmer vs. Snowball Stemmer

**Stemming** is the process of reducing words to their base or root form, often used in natural language processing to handle different word variations. Two popular stemming algorithms are **Porter Stemmer** and **Snowball Stemmer**.

#### **Porter Stemmer**:
- Developed by Martin Porter in 1980.
- One of the most widely used stemming algorithms.
- Works by iteratively applying a set of predefined rules to remove word suffixes.
- Tends to produce stems that may not always be valid words (e.g., *"running"* becomes *"run"* but *"happiness"* becomes *"happi"*).
- Known for its **complexity**, as it involves several phases and exceptions.
- More **aggressive** in reducing words, which may result in over-stemming (producing overly short or inaccurate stems).
- It is relatively slow compared to modern algorithms like Snowball.

#### **Snowball Stemmer**:
- Also developed by Martin Porter, as an improvement over the original Porter Stemmer.
- Part of the **Snowball project**, sometimes referred to as **Porter2**.
- More efficient, faster, and easier to maintain compared to the Porter Stemmer.
- Provides a more **modular** and extensible approach to stemming with clearly defined rules.
- Supports **multiple languages**, unlike the Porter Stemmer, which is mainly for English.
- Produces more **accurate stems**, reducing the likelihood of over-stemming.
- Generally preferred in modern applications due to its speed and accuracy.

#### Summary:
- **Porter Stemmer** is more aggressive and complex, while **Snowball Stemmer** is a refined version offering better performance and multi-language support.


```python
words = ['eating', 'driving', 'eaten', 'collecting', 'collection', 
         'race', 'racing', 'races']

from nltk.stem import PorterStemmer

stemming = PorterStemmer()

for word in words: 
    print(word + " -------> "+ stemming.stem(word))
```

```python
from nltk.stem import SnowballStemmer

Snowball_stemmer = SnowballStemmer("english")
stemmed_words = [Snowball_stemmer.stem(word) for word in words]
for original, stemmed in zip(words, stemmed_words):
    print(original, " -------> ", stemmed)
```

### Stopwords in NLP

**Stopwords** are commonly used words in a language that are often filtered out in natural language processing (NLP) tasks because they carry little semantic meaning. Removing stopwords helps focus on more important words in the text.

#### Examples of Stopwords:
- In English, typical stopwords include: `"the"`, `"is"`, `"in"`, `"and"`, `"of"`, `"to"`, `"it"`, etc.
- These words are frequent in sentences but donâ€™t contribute much to the content or meaning.

#### Importance of Stopwords in NLP:
- **Efficiency**: Removing stopwords reduces the size of text data, which improves processing speed and reduces noise.
- **Relevance**: It helps focus algorithms on the most meaningful words, improving accuracy for tasks like text classification, sentiment analysis, and keyword extraction.
  
#### Common Uses of Stopwords:
- **Text Preprocessing**: In tasks like information retrieval, machine learning, or natural language understanding, stopwords are typically removed during preprocessing.
- **Search Engines**: Search engines often ignore stopwords to provide more relevant results by focusing on keywords.

#### Challenges:
- Some stopwords can be contextually important depending on the use case. For example, in sentiment analysis, words like `"not"` may need to be retained as they can change the meaning of a sentence.

#### Managing Stopwords:
- NLP libraries like **NLTK**, **spaCy**, and **Scikit-learn** provide pre-defined stopword lists that can be customized based on the specific requirements of a task.


```python
from nltk.corpus import stopwords

sw_en = stopwords.words("english")
print(len(sw_en))
print(sw_en[:5], "...", sw_en[-5:])
```

### WordNet

**WordNet** is a large lexical database of the English language, created to support natural language processing (NLP) tasks and linguistic research. It groups words into sets of synonyms called **synsets** and provides short definitions and usage examples. WordNet is widely used in tasks like word sense disambiguation, semantic analysis, and information retrieval.

#### Key Features of WordNet:
- **Synsets (Synonym Sets)**: Words that are grouped based on their meanings. Each synset represents a single concept and contains words that are interchangeable in at least one context.
  - Example: The words `"car"`, `"automobile"`, and `"machine"` are part of the same synset because they all can mean a motor vehicle.
  
- **Hypernyms and Hyponyms**:
  - **Hypernyms**: More general terms (e.g., `"vehicle"` is a hypernym of `"car"`).
  - **Hyponyms**: More specific terms (e.g., `"sedan"` is a hyponym of `"car"`).
  
- **Meronyms and Holonyms**:
  - **Meronyms**: Parts of something (e.g., `"wheel"` is a meronym of `"car"`).
  - **Holonyms**: The whole that something is part of (e.g., `"car"` is a holonym of `"wheel"`).

- **Antonyms**: WordNet also captures antonyms (words with opposite meanings) such as `"hot"` and `"cold"`.

#### Applications of WordNet:
- **Word Sense Disambiguation**: Helps determine the correct meaning of a word in context by linking it to the appropriate synset.
- **Semantic Similarity**: Enables calculation of the similarity between words or concepts based on their synset relations.
- **Information Retrieval**: Improves search results by using semantic information to match synonyms, helping users find relevant documents even if the exact search terms arenâ€™t used.

#### Libraries that Use WordNet:
- **NLTK**: The Python NLTK library provides easy access to WordNet, making it a powerful tool for NLP applications.
- **spaCy**: Another popular NLP library that can integrate with WordNet to enhance its linguistic capabilities.

#### Example Use Case:
In a sentence like `"He is driving a car."`, WordNet can identify that `"car"` belongs to a synset related to motor vehicles and can connect it with similar concepts like `"automobile"`, improving the understanding of the text.


```python
from nltk.corpus import wordnet

# The synsets() method return a list of Synset (synonym set) objects
ss_car_list = wordnet.synsets("motorcar")

# In this example, the word "motorcar" has just one possible meaning and it is identified as "car.n.01"
print(ss_car_list)

for ss_car in ss_car_list:
   print("Synonymous words (or 'lemmas'):", ss_car.lemma_names())
   print("Definition:", ss_car.definition())
   print("Examples:", ss_car.examples())
   print("Types:", ss_car.hyponyms())
   print("Parts:", ss_car.part_meronyms())
```

### Bag of Words (BoW) Model

**Bag of Words (BoW)** is a simple and widely used technique in natural language processing (NLP) for representing text data. It is used to convert text into numerical features that can be used by machine learning algorithms.

#### Key Concepts:

- **Text Representation**:
  - The BoW model represents text as an unordered collection of words, ignoring grammar and word order.
  - Each unique word in the text corpus is treated as a feature.

- **Feature Vector**:
  - In BoW, each document is represented as a vector with dimensions equal to the size of the vocabulary.
  - The vector values typically represent word frequencies or binary indicators (presence or absence of a word).

- **Example**:
  - Suppose we have the following two sentences:
    - Sentence 1: `"The cat sat on the mat."`
    - Sentence 2: `"The dog sat on the log."`
  - The vocabulary for these sentences would be: `["The", "cat", "sat", "on", "the", "mat", "dog", "log"]`
  - The BoW representation of Sentence 1 might look like `[1, 1, 1, 1, 1, 1, 0, 0]` (indicating the presence of the words in the vocabulary).
  - For Sentence 2, it would be `[1, 0, 1, 1, 1, 0, 1, 1]`.


```python
from sklearn.feature_extraction.text import CountVectorizer

corpus = ["Computers can analyze text", 
          "They do it using vectors and matrices", 
          "Computers can process massive amounts of text data"]

vectorizer = CountVectorizer(stop_words="english")
X = vectorizer.fit_transform(corpus)

print("Vocabulary", vectorizer.vocabulary_)
print("Feature names", vectorizer.get_feature_names_out())
print("BoW matrix", X.toarray())
print("BoW matrix shape", X.shape)
```

### TF-IDF (Term Frequency-Inverse Document Frequency)

**TF-IDF** is a statistical measure used to evaluate the importance of a word in a document relative to a collection of documents (corpus). It combines two metrics: Term Frequency (TF) and Inverse Document Frequency (IDF), to reflect how important a word is to a specific document within the corpus.

#### Key Concepts:

- **Term Frequency (TF)**:
  - Measures how frequently a term (word) appears in a document.
  - Formula: `TF(t, d) = (Number of times term t appears in document d) / (Total number of terms in document d)`
  
- **Inverse Document Frequency (IDF)**:
  - Measures how important a term is across the entire corpus, reducing the weight of common terms that appear in many documents.
  - Formula: `IDF(t, D) = log((Total number of documents in corpus D) / (Number of documents containing term t))`
  
- **TF-IDF Score**:
  - The product of TF and IDF, giving a numerical value that represents the importance of the term in the document relative to the corpus.
  - Formula: `TF-IDF(t, d, D) = TF(t, d) * IDF(t, D)`

#### Example:

- Consider a document containing the term `"machine"` and another document containing `"learning"`.
  - If `"machine"` appears frequently in the document but is common across many documents, its TF-IDF score might be lower compared to `"learning"`, which may be rare in the corpus but significant in the specific document.

```python
from sklearn.feature_extraction.text import TfidfVectorizer

vectorizer = TfidfVectorizer(stop_words="english")
X = vectorizer.fit_transform(corpus)

print("Corpus", corpus)
print("Feature names", vectorizer.get_feature_names_out())
print("TF-IDF matrix", X.toarray())
print("TF-IDF matrix shape", X.shape)
```

### Cosine Similarity

**Cosine Similarity** is a metric used to measure how similar two vectors (such as text documents or feature vectors) are, based on their orientation rather than their magnitude. It is commonly used in natural language processing (NLP) and information retrieval to determine the similarity between text documents.

#### Key Concepts:

- **Vector Representation**:
  - Text documents are often represented as vectors in a high-dimensional space, where each dimension corresponds to a term or feature.
  
- **Cosine Similarity Formula**:
  - The cosine similarity between two vectors $\vec{x}$ and $\vec{y}$ is calculated using the following formula:

    $$
    \text{cosine\_similarity}(\vec{x}, \vec{y}) = \frac{\vec{x} \cdot \vec{y}}{\|\vec{x}\| \|\vec{y}\|}
    $$

    Where:

    - $\vec{x} \cdot \vec{y}$ is the dot product of the vectors $\vec{x}$ and $\vec{y}$.
    - $\|\vec{x}\|$ is the magnitude (or length) of vector $\vec{x}$.
    - $\|\vec{y}\|$ is the magnitude (or length) of vector $\vec{y}$.

    The dot product is calculated as:

    $$
    \vec{x} \cdot \vec{y} = \sum_{i=1}^{n} x_i y_i
    $$

    The magnitude of a vector $\vec{x}$ is calculated as:

    $$
    \|\vec{x}\| = \sqrt{\sum_{i=1}^{n} x_i^2}
    $$

    For example, if we have two vectors:

    - $\vec{x} = [3, 2, 0, 5]$ 
    - $\vec{y} = [1, 0, 0, 0]$

    We can calculate cosine similarity as follows:

    1. Calculate the dot product:

    $$
    \vec{x} \cdot \vec{y} = 3*1 + 2*0 + 0*0 + 5*0 = 3
    $$

    2. Calculate the magnitudes:

    $$
    \|\vec{x}\| = \sqrt{3^2 + 2^2 + 0^2 + 5^2} = \sqrt{38} \approx 6.16
    $$
    
    $$  
    \|\vec{y}\| = \sqrt{1^2 + 0^2 + 0^2 + 0^2} = 1
    $$

    3. Finally, compute cosine similarity:

    $$
    \text{cosine\_similarity}(\vec{x}, \vec{y}) = \frac{3}{6.16 * 1} \approx 0.49
    $$

  
- **Range**:
  - The similarity score ranges from -1 to 1, where:
    - **1** indicates that the vectors are identical (perfectly similar).
    - **0** indicates no similarity.
    - **-1** indicates that the vectors are diametrically opposed (inversely similar).

#### Example:

- If two documents are represented as vectors and their cosine similarity is calculated to be 0.85, it means the documents are quite similar in terms of their content.


```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity

vectorizer = TfidfVectorizer(stop_words="english")
X = vectorizer.fit_transform(corpus)

similarity_matrix = cosine_similarity(X)

print(corpus)
print(similarity_matrix)

type(X.toarray())
     
```
